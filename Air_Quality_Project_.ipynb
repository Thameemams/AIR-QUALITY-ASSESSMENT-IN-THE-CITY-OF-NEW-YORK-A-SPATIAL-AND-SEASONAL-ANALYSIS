# **1.Univariate Analysis**

# 1.1.Hist Plot-Distribution of Air Quality Values Recorded in New York City

The seaborn.histplot() function is a versatile tool in the Seaborn Python library used for visualizing the distribution of a dataset by plotting a histogram.

Key uses:

*   Understand Data Shape

*   Analyze Continuous Data
*   Spot Patterns & Outliers


*   Compare Distributions



```python
sns.histplot(df['Data Value'],bins=8,edgecolor="black",kde="True",color="darkblue")
plt.title("Distribution of Air Quality Values")
plt.xlabel("Air Quality Data Value")
plt.ylabel("Count of Values")
plt.show()
```

**Output**:The chart reveals that air quality values are predominantly concentrated between 0 and 50, with a notable peak around 25, and gradually decrease as they approach 100.

# 1.2.Pie Chart-Composition of Air Pollutants in New York City

A pie chart in Matplotlib is a circular statistical graph that uses slices to represent the proportional distribution or percentage of different categories within a whole dataset.

Key uses:

*   To show Part-to-Whole Relationship

*   Clear Proportions

*   Dominant Slices




```python
pollutant_composition = df['Name'].value_counts()
plt.figure(figsize=(10, 12))
plt.pie(pollutant_composition.values,
        labels=pollutant_composition.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors)
plt.title('Composition of Air Pollutants')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

**Output**:In New York City, both nitrogen dioxide and fine particles (PM2.5) each account for 33.6% of total pollution, followed by ozone at 11.2%. Other pollutants are present only in trace amounts.

# 1.3.Bar Plot-Geographic Coverage of  New York City

A Seaborn bar plot (using sns.barplot()) visualizes the relationship between a categorical and a numerical variable, showing the central tendency of the numeric data for each category.

Key uses:

*   To compare discrete categories of data

*   Show rankings
*   Highlight differences

*   Visualize quantities/frequencies for distinct groups








```python
geo_type_counts = df['Geo Type Name'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=geo_type_counts.index, y=geo_type_counts.values)
plt.title('Geographic Coverage of Study')
plt.xlabel('Geographic Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Output**:The 42 United Hospital Fund (UHF) districts cover the majority of the geographical area of New York City, followed by Community Districts, UHF34, Boroughs, and Citywide divisions.

# 2.1.Bar Plot-Most Polluted Places in New York City

 A bar plot is utilized to compare the average pollution levels recorded in various geographic locations.

# 2.3.Box Plot-Pollutants Values by Geographic Area

A Seaborn box plot is a visualization tool used to show the distribution of quantitative data across different categories. It provides a standardized "five-number summary" of a dataset, which helps in understanding data spread, central tendency, and identifying potential outliers.

Here, a box plot is used to visualize pollution trends across different geographic areas.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Data Value', y='Geo Type Name', data=df)
plt.title('Pollutant Values by Geographic Type')
plt.xlabel('Data Value')
plt.ylabel('Geographic Type')
plt.show()
```

**Output**:The highest variability in pollution levels is observed in the 42 United Hospital Fund (UHF) districts, where pollutant values can reach up to 400. # 2.4.Violin Plot-Pollutant Values by Type of Pollutants

Violin plot is used to visualize the distribution of a quantitative variable across one or more categorical variables, providing a richer understanding of the data's shape and spread than a box plot alone.

Violin plot is used to:


* Observe the full shape and probability density of numerical data.  

* Compare distributions across multiple categories, not just summary statistics.  
*  Identify if a dataset has multiple peaks (clusters) or a skewed distribution, which are hard to spot with just a box plot.








```python
plt.figure(figsize=(8, 8))
sns.violinplot(x='Name', y='Data Value', data=df, palette='muted', inner='quartile', hue='Name', legend=False)
plt.title('Distribution of Data Values Across Different Pollutants (Violin Plot)')
plt.xlabel('Pollutant Name')
plt.ylabel('Data Value')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Output**:The health-related indicators, particularly asthma emergency department visits due to PM2.5, exhibit the greatest spread and highest extreme values, peaking at over 400. Fine particles (PM2.5) display a very tight and compact distribution, whereas nitrogen dioxide (NO2) shows a wider and more varied distribution.

# **3.Multivariate Analysis**

Multivariate analysis is a set of statistical methods for analyzing data with multiple variables simultaneously, revealing complex patterns, relationships and structures.

# 3.1.Line Plot-Trend of Top Five Pollutants over Time

A line plot is a data visualization that displays the relationship between two variables, using connected data points to show trends over a continuous range, often time. It is typically used to visualize changes, patterns, and fluctuations in data.

Key uses:

*   To Analyze Trends and Patterns

*   Comparing Multiple Datasets
*   Highlighting Anomalies


*   Forecasting



```python
top_5 = df['Name'].value_counts().nlargest(5).index
df_top5 = df[df['Name'].isin(top_5)]
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_top5, x='Year', y='Data Value', hue='Name')
plt.title('Pollutant Trends Over Time')
plt.show()
```

**Output**:In New York City, asthma emergency department visits due to PM2.5 represent the largest health-related impact of pollution, with the trend showing a decrease from 2006 to 2017. In contrast, the other leading indicators display either increasing or static trends from 2007 to 2022.

# 3.2.Heatmap-Mean Pollutant Level by Geographic Area

Heatmap displays rectangular data (like matrices or correlation tables) as a colour-coded grid, using colour intensity and hue to represent data values, making patterns, correlations, and outliers instantly visible, ideal for showing relationships between variables or trends across dimensions.

Key uses:


*   To visualize complex data

*   To spot trends/hotspots
*   To  understand user behavior by showing variations with colour


*   To see highs and lows across multiple variables at a glance



```python
pivot_df = df.pivot_table(index='Name', columns='Geo Type Name',observed=True, values='Data Value', aggfunc='mean')
plt.figure(figsize=(12, 10))
sns.heatmap(pivot_df, annot=True, cmap='YlGnBu')
plt.title('Mean Pollutant Levels by Geography Type')
plt.xlabel('Geographic Type')
plt.ylabel('Pollutant Name')
plt.show()
```

The highest numerical values in the dataset are found in annual vehicle miles traveled, which remain consistently high across Community Districts (CD) and UHF42 neighborhoods. Both UHF42 and CD generally exhibit higher mean values for primary pollutants such as nitrogen dioxide (NO2) and fine particles (PM2.5). This indicates that pollution is more concentrated at the neighborhood level, where traffic and local emissions (such as from boilers) are most prevalent.
